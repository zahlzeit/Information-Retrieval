```python
#This is the initial crawling

final = pd.DataFrame()
mywebPage = requests.get('https://pureportal.coventry.ac.uk/en/organisations/centre-for-intelligent-healthcare/publications/').text
soup1 = BeautifulSoup(mywebPage, 'lxml')

page_no = []
for i in soup1.find_all('a', class_='step'):
    page_no.append(i.text.strip())

number_of_pages = int(page_no[-1])

counter = 0
for i in range(number_of_pages):
    print(i)
    webPage = requests.get(url = 'https://pureportal.coventry.ac.uk/en/organisations/centre-for-intelligent-healthcare/publications/?page={}'.format(i)).text
    soup = BeautifulSoup(webPage, 'lxml')
    publication = soup.find_all('li', class_='list-result-item')
    print("Done Loop: " + str(i))
    
    SN = []
    title = []
    link = []
    authors = []
    date = []
    for item in publication:
        ititle = item.find('a', class_='link') 
        title.append(ititle.text.strip())

        ilink = item.find('a', href=True)['href']
        link.append(ilink)
        idate = item.find('span', class_='date').text.strip()

        date.append(idate)

        auth_list = []
        alist = item.find_all('a', class_='link person')
        for a in alist:
            auth_list.append(a.text.strip())

        for i in item.find('h3', class_='title').find_next_siblings('span'):
            if "Jan" in i.text.strip() or "Feb" in i.text.strip() or "Mar" in i.text.strip() or "Apr" in i.text.strip() or "May" in i.text.strip() or "Jun" in i.text.strip() or "Jul" in i.text.strip() or "Aug" in i.text.strip() or "Sept" in i.text.strip() or "Oct" in i.text.strip() or "Nov" in i.text.strip() or "Dec" in i.text.strip() or "19" in i.text.strip() or "20" in i.text.strip():
                break
            else:
                auth_list.append(i.text.strip())
        authors.append(auth_list)
        counter = counter + 1
        SN.append(counter)
    d={'SN':SN, 'Title':title, 'Authors':authors, 'Date':date, 'Link':link}
    df=pd.DataFrame(d)
    
    final=final.append(df)
```

```python
# Remove stop words
sw = stopwords.words("english")
lemmatizer = WordNetLemmatizer()


def tp1(txt):
    txt = txt.lower()   # Make lowercase
    txt = txt.translate(str.maketrans('',
                                      '',
                                      string.punctuation))   # Remove punctuation marks
    txt = lematize(txt)
    return txt

def fwpt(word):
    tag = pos_tag([word])[0][1][0].upper()
    hash_tag = {"V": wordnet.VERB, "R": wordnet.ADV,"N": wordnet.NOUN,"J": wordnet.ADJ}         
    return hash_tag.get(tag, wordnet.NOUN)

def lematize(text):
        tkns = nltk.word_tokenize(text)
        ax = ""
        for each in tkns:
            if each not in sw:
                ax += lemmatizer.lemmatize(each, fwpt(each)) + " "
        return ax
```

```python
## Indexer Function
def apply_index(inputs, index):
    words = inputs.Title.split()
    SN = int(inputs.SN)
    for word in words:
        if word in index.keys():
            if SN not in index[word]:
                index[word].append(SN)
        else:
            index[word] = [SN]
    return index

indx = apply_index(inputs=single, index= {})
```

```python
def full_index(df, index):
    for x in range(len(df)):
        inpt = df.iloc[x,:]
        ind = apply_index(inputs=inpt, index=index)
    return ind

def construct_index(df, index):
    queue = preprocess_df(df)
    ind = full_index(df=queue, index=index)
    return ind

indexed = full_index(processed_db, 
                     index = {})


indexes = construct_index(df=final, 
                          index = {})
```

```python
#Or Function
def union(lists):
    union = list(set.union(*map(set, lists)))
    union.sort()
    return union

#And Function
def intersection(lists):
    intersect = list(set.intersection(*map(set, lists)))
    intersect.sort()
    return intersect
```

```python
#Search Function

def vertical_search_engine(df, query, index=indexes):
    query_split = split_query(query)
    retrieved = []
    for word in query_split:
        if word in index.keys():
            retrieved.append(index[word])
            
            
    # Ranked Retrieval
    if len(retrieved)>0:
        high_rank_result = intersection(retrieved)
        low_rank_result = union(retrieved) 
        c = [x for x in low_rank_result if x not in high_rank_result]      
        high_rank_result.extend(c)
        result = high_rank_result
        
        final_output = df[df.SN.isin(result)].reset_index(drop=True)
    
        # Return result in order of Intersection ----> Union
        dummy = pd.Series(result, name = 'SN').to_frame()
        result = pd.merge(dummy, final_output, on='SN', how = 'left')
        
    else:
        result = 'No result found'
    
    return result
```
